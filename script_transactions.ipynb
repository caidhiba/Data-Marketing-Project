```python
import pandas as pd

try:
    df = pd.read_csv('/content/transactions.csv')
    print("Dataset 'transactions.csv' loaded successfully.")
except FileNotFoundError:
    print("dataset not found")
    df = None

if df is not None:

    if 'invoice_id' in df.columns:
        unique_invoices = df['invoice_id'].nunique()
        print(f"\nNumber of unique invoice_id: {unique_invoices}")
    else:
        print("column not found")

    if 'customer_id' in df.columns:
        unique_customers = df['customer_id'].nunique()
        print(f"Number of unique customer_id: {unique_customers}")
    else:
        print("column not found")
```

```python
if df is not None:
    if 'invoice_date' in df.columns:
        try:
            df['invoice_date'] = pd.to_datetime(df['invoice_date'])
            oldest_date = df['invoice_date'].min()
            most_recent_date = df['invoice_date'].max()
            print(f"\nOldest invoice_date: {oldest_date}")
            print(f"Most recent invoice_date: {most_recent_date}")
        except Exception as e:
            print(f"Error converting 'invoice_date' to datetime: {e}")
    else:
        print("Column 'invoice_date' not found in the dataset.")
else:
    print("DataFrame 'df' is not loaded.")
```

```python
if 'oldest_date' in locals() and 'most_recent_date' in locals():
    time_difference = most_recent_date - oldest_date
    print(f"différence temporelle = {time_difference}")
else:
    print("erreur")
```

```python
if df is not None:
    print("\nNombre de valeurs manquantes par colonne :")
    print(df.isnull().sum())
else:
    print("DataFrame 'df' n'est pas chargé. Veuillez charger le dataset d'abord.")
```

```python
if df is not None:
    if 'quantity' in df.columns:
        # Filter rows where 'quantity' is null
        empty_quantity_rows = df[df['quantity'].isnull()]

        if not empty_quantity_rows.empty:
            print("\nLignes avec une quantité vide :")
            print(empty_quantity_rows)
        else:
            print("\nAucune ligne avec une quantité vide n'a été trouvée.")
    else:
        print("\nLa colonne 'quantity' n'existe pas dans le DataFrame.")
else:
    print("\nLe DataFrame 'df' n'est pas chargé.")
```

```python
if df is not None:
    # Count total duplicate rows
    total_duplicates = df.duplicated().sum()
    print(f"\nNombre total de lignes en double: {total_duplicates}")

    if total_duplicates > 0:
        # Display all duplicate rows
        duplicate_rows = df[df.duplicated(keep=False)]
        print("\nLignes en double (affichage complet) :")
        print(duplicate_rows.sort_values(by=list(df.columns)))
    else:
        print("\nAucune ligne en double n'a été trouvée dans le DataFrame.")
else:
    print("\nLe DataFrame 'df' n'est pas chargé.")
```

```python
if df is not None:
    df_cleaned = df.drop_duplicates()
    remaining_rows = df_cleaned.shape[0]
    print(f"\nAprès suppression des doublons, le nombre total de lignes est de : {remaining_rows}")
else:
    print("\nLe DataFrame 'df' n'est pas chargé.")
```
